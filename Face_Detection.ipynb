# Face Detection

```python
face_cascade = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
```

```python
def detect_face(img):
    
    face_img = img.copy()
    
    face_rect = face_cascade.detectMultiScale(face_img)
    
    for (x, y, w, h) in face_rect:
        
        cv2.rectangle(face_img,(x,y),(x+w,y+h),(255,0,0),10)
    
    return face_img
```

```python
img1 = detect_face(face)
```

```python
plt.imshow(img1, cmap='gray')
```

```python
img2 = detect_face(conf)
```

```python
plt.imshow(img2, cmap='gray')
```

```python
def detect_face2(img):
    
    face_img = img.copy()
    
    face_rect = face_cascade.detectMultiScale(face_img, scaleFactor=1.2, minNeighbors=5)
    
    for (x, y, w, h) in face_rect:
        
        cv2.rectangle(face_img,(x,y),(x+w,y+h),(255,0,0),10)
    
    return face_img
```

```python
img3 = detect_face2(conf)
```

```python
plt.imshow(img2, cmap='gray')
```

# Real Time Face Detection

```python
cap = cv2.VideoCapture(0)

while True:

    ret, frame = cap.read()
    
    frame = detect_face2(frame)
    
    cv2.imshow("Camera", frame)
    
    c = cv2.waitKey(1)
    
    if c == 27:
        break
        
cap.release()

cv2.destroyAllWindows()
```

# Eye Detection

```python
eye_cascade = cv2.CascadeClassifier('haarcascade_eye.xml')
def detect_eye(img):
    
    face_img = img.copy()
    
    face_rect = eye_cascade.detectMultiScale(face_img, scaleFactor=1.2, minNeighbors=5)
    
    for (x, y, w, h) in face_rect:
       
        cv2.rectangle(face_img,(x,y),(x+w,y+h),(255,0,0),6)
    
    return face_img
```

```python
img4 = detect_eye(face)
```

```python
plt.imshow(img4)
```

# Real Time Eye Detection

```python
cap = cv2.VideoCapture(0)

while True:
    
    ret, frame = cap.read()
    
    frame = detect_eye(frame)
    
    cv2.imshow("Camera", frame)
    
    c = cv2.waitKey(1)
    
    if c == 27:
        break

        cap.release()
cv2.destroyAllWindows()
```
